```python
# Análise Automática de Planilhas com IA
# Este código permite fazer uma análise exploratória automática de dados de negócios
# e gerar insights relevantes para tomada de decisão, além de um relatório final com OpenAI

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
from google.colab import files
import io
import json
import requests
from google.colab import userdata
import time

# Configurações para visualização
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
plt.style.use('ggplot')

def carregar_dados():
    """
    Solicita o upload de um arquivo CSV ou Excel e carrega os dados

    Returns:
        DataFrame: Dados carregados da planilha
    """
    print("Por favor, faça upload do arquivo CSV ou Excel para análise...")
    uploaded = files.upload()

    # Obter o nome do arquivo enviado
    filename = next(iter(uploaded))

    # Carregar o arquivo com base na extensão
    if filename.endswith('.csv'):
        # Tentar diferentes delimitadores e encodings
        try:
            df = pd.read_csv(io.BytesIO(uploaded[filename]))
        except:
            try:
                df = pd.read_csv(io.BytesIO(uploaded[filename]), sep=';')
            except:
                try:
                    df = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='latin1')
                except:
                    df = pd.read_csv(io.BytesIO(uploaded[filename]), sep=';', encoding='latin1')
    elif filename.endswith(('.xlsx', '.xls')):
        df = pd.read_excel(io.BytesIO(uploaded[filename]))
    else:
        raise ValueError("Formato de arquivo não suportado. Use .csv, .xlsx ou .xls")

    print(f"Arquivo '{filename}' carregado com sucesso!")
    return df, filename

def analise_exploratoria(df):
    """
    Realiza uma análise exploratória automática dos dados

    Args:
        df (DataFrame): DataFrame a ser analisado

    Returns:
        dict: Informações sobre as colunas por tipo
    """
    # Informações gerais do dataset
    print("="*80)
    print("VISÃO GERAL DO DATASET")
    print("="*80)
    print(f"Número de linhas: {df.shape[0]}")
    print(f"Número de colunas: {df.shape[1]}")
    print("\nPrimeiras 5 linhas:")
    print(df.head())

    # Análise de tipos de dados e valores ausentes
    print("\n" + "="*80)
    print("TIPOS DE DADOS E VALORES AUSENTES")
    print("="*80)
    tipos = pd.DataFrame(df.dtypes, columns=['Tipo de Dado'])
    missing = pd.DataFrame(df.isnull().sum(), columns=['Valores Ausentes'])
    missing_percent = pd.DataFrame(df.isnull().sum() / len(df) * 100, columns=['% Ausentes'])
    info_colunas = pd.concat([tipos, missing, missing_percent], axis=1)
    print(info_colunas)

    # Estatísticas descritivas para colunas numéricas
    colunas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
    if len(colunas_numericas) > 0:
        print("\n" + "="*80)
        print("ESTATÍSTICAS DESCRITIVAS (COLUNAS NUMÉRICAS)")
        print("="*80)
        stats = df[colunas_numericas].describe().T
        stats['cv'] = stats['std'] / stats['mean']  # Coeficiente de variação
        print(stats)

        # Verificação de correlações
        print("\n" + "="*80)
        print("MATRIZ DE CORRELAÇÃO")
        print("="*80)
        corr = df[colunas_numericas].corr()
        print(corr)

        # Identificação das correlações mais fortes (positivas e negativas)
        print("\n" + "="*80)
        print("CORRELAÇÕES MAIS FORTES")
        print("="*80)
        corr_unstack = corr.unstack()
        corr_unstack = corr_unstack[corr_unstack < 1.0]  # Remove a diagonal
        top_corr = corr_unstack.abs().sort_values(ascending=False).head(10)
        print(top_corr)

    # Análise de colunas categóricas
    colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns
    if len(colunas_categoricas) > 0:
        print("\n" + "="*80)
        print("ANÁLISE DE COLUNAS CATEGÓRICAS")
        print("="*80)
        for coluna in colunas_categoricas:
            print(f"\nDistribuição da coluna: {coluna}")
            contagem = df[coluna].value_counts()
            percentual = df[coluna].value_counts(normalize=True) * 100
            cat_info = pd.DataFrame({'Contagem': contagem, 'Percentual (%)': percentual})
            print(cat_info.head(10))  # Mostra apenas as 10 categorias mais frequentes
            print(f"Número total de categorias: {len(contagem)}")

    # Detecção automática de datas
    possivel_data = []
    for coluna in df.columns:
        if df[coluna].dtype == 'object':
            try:
                pd.to_datetime(df[coluna], errors='raise')
                possivel_data.append(coluna)
            except:
                pass

    if possivel_data:
        print("\n" + "="*80)
        print("POSSÍVEIS COLUNAS DE DATA IDENTIFICADAS")
        print("="*80)
        print(possivel_data)

        # Converter as colunas de data para o tipo datetime
        for coluna in possivel_data:
            df[coluna] = pd.to_datetime(df[coluna])

    return {
        'numericas': list(colunas_numericas),
        'categoricas': list(colunas_categoricas),
        'datas': possivel_data
    }

def visualizar_distribuicoes_numericas(df, colunas_numericas, max_colunas=5):
    """
    Visualiza a distribuição das variáveis numéricas

    Args:
        df (DataFrame): DataFrame com os dados
        colunas_numericas (list): Lista de colunas numéricas
        max_colunas (int): Número máximo de colunas para visualizar
    """
    print("\n" + "="*80)
    print("DISTRIBUIÇÃO DE VARIÁVEIS NUMÉRICAS")
    print("="*80)

    # Limitar o número de colunas para visualização
    colunas_para_visualizar = colunas_numericas[:min(max_colunas, len(colunas_numericas))]

    for coluna in colunas_para_visualizar:
        fig, ax = plt.subplots(1, 2, figsize=(15, 5))

        # Histograma
        sns.histplot(df[coluna].dropna(), kde=True, ax=ax[0])
        ax[0].set_title(f'Distribuição de {coluna}')

        # Boxplot
        sns.boxplot(x=df[coluna].dropna(), ax=ax[1])
        ax[1].set_title(f'Boxplot de {coluna}')

        plt.tight_layout()
        plt.show()

        # Estatísticas descritivas da coluna
        print(f"\nEstatísticas de {coluna}:")
        stats = df[coluna].describe()
        print(stats)

        # Verificar outliers
        Q1 = df[coluna].quantile(0.25)
        Q3 = df[coluna].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[coluna] < Q1 - 1.5 * IQR) | (df[coluna] > Q3 + 1.5 * IQR)]
        print(f"Número de possíveis outliers: {len(outliers)}")
        print(f"Percentual de outliers: {len(outliers) / len(df) * 100:.2f}%")

def visualizar_distribuicoes_categoricas(df, colunas_categoricas, max_colunas=5):
    """
    Visualiza a distribuição das variáveis categóricas

    Args:
        df (DataFrame): DataFrame com os dados
        colunas_categoricas (list): Lista de colunas categóricas
        max_colunas (int): Número máximo de colunas para visualizar
    """
    print("\n" + "="*80)
    print("DISTRIBUIÇÃO DE VARIÁVEIS CATEGÓRICAS")
    print("="*80)

    # Limitar o número de colunas para visualização
    colunas_para_visualizar = colunas_categoricas[:min(max_colunas, len(colunas_categoricas))]

    for coluna in colunas_para_visualizar:
        # Verificar número de categorias
        num_categorias = df[coluna].nunique()

        if num_categorias > 20:
            print(f"\nA coluna {coluna} tem {num_categorias} categorias únicas. Mostrando apenas as 20 mais frequentes.")
            top_cats = df[coluna].value_counts().head(20).index
            # Criar uma nova série com apenas as top categorias
            serie_filtrada = df[coluna].copy()
            serie_filtrada[~df[coluna].isin(top_cats)] = 'Outras'

            plt.figure(figsize=(12, 6))
            ax = sns.countplot(y=serie_filtrada, order=serie_filtrada.value_counts().index)
            plt.title(f'Distribuição de {coluna} (Top 20 + Outras)')
            plt.xlabel('Contagem')
            plt.ylabel(coluna)

            # Adicionar valores às barras
            for p in ax.patches:
                width = p.get_width()
                plt.text(width + 10, p.get_y() + p.get_height()/2, f'{width:.0f}', ha='left', va='center')

            plt.tight_layout()
            plt.show()
        else:
            plt.figure(figsize=(12, 6))
            ax = sns.countplot(y=df[coluna], order=df[coluna].value_counts().index)
            plt.title(f'Distribuição de {coluna}')
            plt.xlabel('Contagem')
            plt.ylabel(coluna)

            # Adicionar valores às barras
            for p in ax.patches:
                width = p.get_width()
                plt.text(width + 5, p.get_y() + p.get_height()/2, f'{width:.0f}', ha='left', va='center')

            plt.tight_layout()
            plt.show()

        # Estatísticas da coluna
        value_counts = df[coluna].value_counts()
        value_percent = df[coluna].value_counts(normalize=True) * 100

        print(f"\nEstatísticas de {coluna}:")
        print(f"Número de categorias únicas: {num_categorias}")
        print(f"Categoria mais frequente: {value_counts.index[0]} ({value_percent[0]:.2f}%)")
        print(f"Categoria menos frequente: {value_counts.index[-1]} ({value_percent[-1]:.2f}%)")

def visualizar_series_temporais(df, colunas_data, colunas_numericas, max_colunas=5):
    """
    Visualiza séries temporais se dados de data estiverem disponíveis

    Args:
        df (DataFrame): DataFrame com os dados
        colunas_data (list): Lista de colunas de data
        colunas_numericas (list): Lista de colunas numéricas
        max_colunas (int): Número máximo de séries para visualizar
    """
    if not colunas_data or not colunas_numericas:
        return

    print("\n" + "="*80)
    print("ANÁLISE DE SÉRIES TEMPORAIS")
    print("="*80)

    # Selecionar a primeira coluna de data
    coluna_data = colunas_data[0]
    print(f"Usando coluna de data: {coluna_data}")

    # Garantir que a coluna é do tipo datetime
    df[coluna_data] = pd.to_datetime(df[coluna_data])

    # Selecionar as colunas numéricas para análise (limitando ao máximo)
    colunas_para_analise = colunas_numericas[:min(max_colunas, len(colunas_numericas))]

    for coluna in colunas_para_analise:
        # Agrupar por data (mês) e calcular média
        df_agrupado = df.copy()
        df_agrupado['Mês'] = df_agrupado[coluna_data].dt.to_period('M')
        serie_mensal = df_agrupado.groupby('Mês')[coluna].mean()

        # Calcular o mês com valor mais alto e mais baixo
        mes_max = serie_mensal.idxmax()
        mes_min = serie_mensal.idxmin()

        # Calcular tendência (última metade vs primeira metade)
        metade = len(serie_mensal) // 2
        if metade > 0:
            primeira_metade = serie_mensal.iloc[:metade].mean()
            segunda_metade = serie_mensal.iloc[metade:].mean()
            variacao_percentual = ((segunda_metade - primeira_metade) / primeira_metade * 100) if primeira_metade > 0 else 0

            if variacao_percentual > 5:
                tendencia = f"Em alta (+{variacao_percentual:.2f}%)"
            elif variacao_percentual < -5:
                tendencia = f"Em queda ({variacao_percentual:.2f}%)"
            else:
                tendencia = f"Estável ({variacao_percentual:.2f}%)"
        else:
            tendencia = "Dados insuficientes para análise de tendência"

        # Plotar a série temporal
        plt.figure(figsize=(14, 7))
        serie_mensal.plot(marker='o')
        plt.title(f'Evolução de {coluna} ao longo do tempo')
        plt.ylabel(coluna)
        plt.xlabel('Mês')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

        print(f"\nAnálise temporal de {coluna}:")
        print(f"Tendência: {tendencia}")
        print(f"Mês com maior valor: {mes_max} ({serie_mensal[mes_max]:.2f})")
        print(f"Mês com menor valor: {mes_min} ({serie_mensal[mes_min]:.2f})")

def visualizar_correlacoes(df, colunas_numericas):
    """
    Visualiza a matriz de correlação entre variáveis numéricas

    Args:
        df (DataFrame): DataFrame com os dados
        colunas_numericas (list): Lista de colunas numéricas
    """
    if len(colunas_numericas) < 2:
        print("\nNão há colunas numéricas suficientes para análise de correlação.")
        return

    print("\n" + "="*80)
    print("MATRIZ DE CORRELAÇÃO")
    print("="*80)

    # Calcular a matriz de correlação
    corr_matrix = df[colunas_numericas].corr()

    # Visualizar com heatmap
    plt.figure(figsize=(12, 10))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    sns.heatmap(corr_matrix, mask=mask, cmap=cmap, vmax=1, vmin=-1, center=0,
                square=True, linewidths=.5, annot=True, fmt=".2f", cbar_kws={"shrink": .5})
    plt.title('Matriz de Correlação entre Variáveis Numéricas')
    plt.tight_layout()
    plt.show()

    # Identificar correlações mais fortes (positivas e negativas)
    corr_unstack = corr_matrix.unstack()
    corr_unstack = corr_unstack[corr_unstack < 1.0]  # Remove a diagonal

    # Top 5 correlações positivas
    top_pos = corr_unstack[corr_unstack > 0].sort_values(ascending=False).head(5)

    # Top 5 correlações negativas
    top_neg = corr_unstack[corr_unstack < 0].sort_values().head(5)

    print("\nCorrelações positivas mais fortes:")
    for idx, value in top_pos.items():
        print(f"{idx[0]} & {idx[1]}: {value:.4f}")

    print("\nCorrelações negativas mais fortes:")
    for idx, value in top_neg.items():
        print(f"{idx[0]} & {idx[1]}: {value:.4f}")

    # Visualizar scatter plots para as correlações mais fortes
    if len(top_pos) > 0:
        print("\nScatter plots das correlações positivas mais fortes:")
        for idx, value in top_pos.items():
            var1, var2 = idx
            plt.figure(figsize=(10, 6))
            sns.scatterplot(x=df[var1], y=df[var2], alpha=0.6)
            plt.title(f'Correlação entre {var1} e {var2}: {value:.4f}')
            plt.xlabel(var1)
            plt.ylabel(var2)
            plt.tight_layout()
            plt.show()

    if len(top_neg) > 0:
        print("\nScatter plots das correlações negativas mais fortes:")
        for idx, value in top_neg.items():
            var1, var2 = idx
            plt.figure(figsize=(10, 6))
            sns.scatterplot(x=df[var1], y=df[var2], alpha=0.6)
            plt.title(f'Correlação entre {var1} e {var2}: {value:.4f}')
            plt.xlabel(var1)
            plt.ylabel(var2)
            plt.tight_layout()
            plt.show()

def analise_bivariada(df, colunas_numericas, colunas_categoricas, max_analises=3):
    """
    Realiza análises bivariadas entre variáveis numéricas e categóricas

    Args:
        df (DataFrame): DataFrame com os dados
        colunas_numericas (list): Lista de colunas numéricas
        colunas_categoricas (list): Lista de colunas categóricas
        max_analises (int): Número máximo de análises para cada tipo
    """
    if not colunas_numericas or not colunas_categoricas:
        return

    print("\n" + "="*80)
    print("ANÁLISE BIVARIADA (NUMÉRICA VS CATEGÓRICA)")
    print("="*80)

    # Limitar o número de análises
    colunas_num = colunas_numericas[:min(max_analises, len(colunas_numericas))]
    colunas_cat = colunas_categoricas[:min(max_analises, len(colunas_categoricas))]

    for col_num in colunas_num:
        for col_cat in colunas_cat:
            # Verificar se a coluna categórica não tem muitas categorias
            n_categorias = df[col_cat].nunique()

            if n_categorias <= 10:  # Limitar para não ter muitas barras
                plt.figure(figsize=(14, 7))

                # Boxplot agrupado por categoria
                ax = sns.boxplot(x=col_cat, y=col_num, data=df)
                plt.title(f'{col_num} por {col_cat}')
                plt.xlabel(col_cat)
                plt.ylabel(col_num)
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

                # Estatísticas por grupo
                stats_grupo = df.groupby(col_cat)[col_num].agg(['mean', 'median', 'std', 'min', 'max', 'count'])
                print(f"\nEstatísticas de {col_num} por {col_cat}:")
                print(stats_grupo)

                # Verificar se há diferenças significativas entre os grupos
                max_media = stats_grupo['mean'].max()
                min_media = stats_grupo['mean'].min()
                variacao = (max_media - min_media) / min_media * 100 if min_media > 0 else 0

                print(f"\nVariação entre grupos: {variacao:.2f}%")
                if variacao > 20:
                    print(f"INSIGHT: Há uma diferença significativa de {col_num} entre diferentes categorias de {col_cat}.")
                    grupo_max = stats_grupo['mean'].idxmax()
                    grupo_min = stats_grupo['mean'].idxmin()
                    print(f"O grupo '{grupo_max}' tem em média {variacao:.2f}% mais {col_num} que o grupo '{grupo_min}'.")
            else:
                print(f"\nA coluna {col_cat} tem muitas categorias ({n_categorias}). Pulando análise bivariada.")

def identificar_clusters(df, colunas_numericas, n_clusters=3):
    """
    Identifica clusters (agrupamentos) nos dados usando K-means

    Args:
        df (DataFrame): DataFrame com os dados
        colunas_numericas (list): Lista de colunas numéricas
        n_clusters (int): Número de clusters a identificar
    """
    if len(colunas_numericas) < 2:
        print("\nNão há colunas numéricas suficientes para análise de clusters.")
        return

    print("\n" + "="*80)
    print("ANÁLISE DE CLUSTERS")
    print("="*80)

    # Selecionar apenas colunas numéricas e remover linhas com valores ausentes
    df_cluster = df[colunas_numericas].dropna()

    if len(df_cluster) < 10:
        print("Dados insuficientes para clustering após remoção de valores ausentes.")
        return

    # Padronizar os dados
    scaler = StandardScaler()
    df_scaled = scaler.fit_transform(df_cluster)

    # Aplicar K-means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(df_scaled)

    # Adicionar a coluna de cluster ao dataframe original
    df_resultado = df_cluster.copy()
    df_resultado['Cluster'] = clusters

    # Calcular as médias de cada cluster
    medias_cluster = df_resultado.groupby('Cluster').mean()

    print(f"Análise de clusters (K-means, n_clusters={n_clusters})")
    print("\nMédia das variáveis por cluster:")
    print(medias_cluster)

    print("\nContagem de elementos por cluster:")
    print(df_resultado['Cluster'].value_counts())

    # Visualização 2D dos clusters se houver pelo menos 2 dimensões
    if len(colunas_numericas) >= 2:
        plt.figure(figsize=(12, 8))

        # Scatter plot com as duas primeiras variáveis
        plt.subplot(1, 2, 1)
        sns.scatterplot(x=df_resultado[colunas_numericas[0]],
                       y=df_resultado[colunas_numericas[1]],
                       hue=df_resultado['Cluster'],
                       palette='viridis')
        plt.title(f'Clusters visualizados por {colunas_numericas[0]} vs {colunas_numericas[1]}')

        # Gráfico de barras com média por cluster
        plt.subplot(1, 2, 2)
        medias_cluster.T.plot(kind='bar')
        plt.title('Média das variáveis por cluster')
        plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()

        # Características de cada cluster
        print("\nCaracterísticas de cada cluster:")
        for i in range(n_clusters):
            print(f"\nCluster {i}:")
            # Para cada variável, verificar se está acima ou abaixo da média global
            for col in colunas_numericas:
                media_global = df_cluster[col].mean()
                media_cluster = medias_cluster.loc[i, col]
                diff_percent = (media_cluster - media_global) / media_global * 100 if media_global > 0 else 0

                if diff_percent > 20:
                    print(f"- {col} ALTO: {media_cluster:.2f} (+{diff_percent:.1f}% acima da média)")
                elif diff_percent < -20:
                    print(f"- {col} BAIXO: {media_cluster:.2f} ({diff_percent:.1f}% abaixo da média)")
                else:
                    print(f"- {col} MÉDIO: {media_cluster:.2f} ({diff_percent:.1f}% da média)")

def gerar_insights(df, info_colunas):
    """
    Gera insights automáticos baseados nos dados

    Args:
        df (DataFrame): DataFrame com os dados
        info_colunas (dict): Informações sobre as colunas por tipo

    Returns:
        list: Lista de insights encontrados
    """
    print("\n" + "="*80)
    print("INSIGHTS AUTOMÁTICOS")
    print("="*80)

    insights = []

    # Insight 1: Identificação de outliers em variáveis numéricas
    for col in info_colunas['numericas']:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        outliers = df[(df[col] < q1 - 1.5 * iqr) | (df[col] > q3 + 1.5 * iqr)]

        if len(outliers) > 0:
            percentual = len(outliers) / len(df) * 100
            insights.append(f"A variável '{col}' possui {len(outliers)} outliers ({percentual:.2f}% dos dados).")

    # Insight 2: Correlações fortes entre variáveis numéricas
    if len(info_colunas['numericas']) >= 2:
        corr_matrix = df[info_colunas['numericas']].corr()
        for i in range(len(corr_matrix.columns)):
            for j in range(i):
                if abs(corr_matrix.iloc[i, j]) > 0.7:
                    insights.append(f"Correlação forte ({corr_matrix.iloc[i, j]:.2f}) entre '{corr_matrix.columns[i]}' e '{corr_matrix.columns[j]}'.")

    # Insight 3: Distribuição desbalanceada em variáveis categóricas
    for col in info_colunas['categoricas']:
        top_cat = df[col].value_counts(normalize=True).max() * 100
        if top_cat > 80:
            insights.append(f"A categoria mais frequente de '{col}' representa {top_cat:.2f}% dos dados, indicando possível desbalanceamento.")

    # Insight 4: Análise de tendências em dados temporais
    if info_colunas['datas'] and info_colunas['numericas']:
        data_col = info_colunas['datas'][0]
        df_temp = df.copy()
        df_temp[data_col] = pd.to_datetime(df_temp[data_col])
        df_temp = df_temp.sort_values(data_col)

        for col in info_colunas['numericas']:
            # Calcula a tendência (positiva ou negativa)
            try:
                first_half = df_temp.iloc[:len(df_temp)//2][col].mean()
                second_half = df_temp.iloc[len(df_temp)//2:][col].mean()
                percent_change = (second_half - first_half) / first_half * 100 if first_half > 0 else 0

                if abs(percent_change) > 10:  # Alteração significativa
                    direction = "aumento" if percent_change > 0 else "redução"
                    insights.append(f"Tendência de {direction} de {abs(percent_change):.2f}% em '{col}' ao longo do período analisado.")
            except:
                pass

    # Insight 5: Identificação de valores ausentes significativos
    missing_cols = df.columns[df.isnull().mean() > 0.1]
    for col in missing_cols:
        missing_percent = df[col].isnull().mean() * 100
        insights.append(f"A variável '{col}' possui {missing_percent:.2f}% de valores ausentes, o que pode impactar análises.")

    # Insight 6: Análise de variáveis categóricas vs numéricas
    if info_colunas['categoricas'] and info_colunas['numericas']:
        for cat_col in info_colunas['categoricas'][:2]:  # Limitar a 2 colunas categóricas
            if df[cat_col].nunique() < 10:  # Apenas para colunas com poucas categorias
                for num_col in info_colunas['numericas'][:3]:  # Limitar a 3 colunas numéricas
                    try:
                        group_means = df.groupby(cat_col)[num_col].mean()
                        max_cat = group_means.idxmax()
                        min_cat = group_means.idxmin()
                        max_val = group_means.max()
                        min_val = group_means.min()

                        if min_val > 0:
                            diff_percent = (max_val - min_val) / min_val * 100
                            if diff_percent > 30:  # Diferença significativa
                                insights.append(f"A categoria '{max_cat}' de '{cat_col}' tem um valor médio de '{num_col}' {diff_percent:.2f}% maior que a categoria '{min_cat}'.")
                    except:
                        pass

    # Exibir todos os insights encontrados
    if len(insights) > 0:
        print(f"Foram encontrados {len(insights)} insights nos dados:")
        for i, insight in enumerate(insights, 1):
            print(f"{i}. {insight}")
    else:
        print("Não foram encontrados insights significativos nos dados.")

    return insights

def gerar_relatorio_openai(filename, df, info_colunas, insights):
    """
    Envia os dados da análise para a OpenAI e gera um relatório detalhado

    Args:
        filename (str): Nome do arquivo analisado
        df (DataFrame): DataFrame com os dados
        info_colunas (dict): Informações sobre as colunas por tipo
        insights (list): Lista de insights encontrados

    Returns:
        str: Relatório gerado pela OpenAI
    """
    print("\n" + "="*80)
    print("GERANDO RELATÓRIO DE ANÁLISE COM OPENAI")
    print("="*80)

    try:
        # Obter a API key do secrets do Google Colab
        api_key = userdata.get('OPENAI_API_KEY')
        if not api_key:
            raise ValueError("API key da OpenAI não encontrada nos secrets do Colab.")

        # Preparar o prompt para enviar à OpenAI
        prompt = f"""
        Você é um analista de dados especializado. Preciso que você analise os dados abaixo e forneça um relatório detalhado.

        Informações gerais:
        - Nome do arquivo: {filename}
        - Número de linhas: {len(df)}
        - Número de colunas: {len(df.columns)}

        Tipos de colunas:
        - Colunas numéricas: {', '.join(info_colunas['numericas']) if info_colunas['numericas'] else 'Nenhuma'}
        - Colunas categóricas: {', '.join(info_colunas['categoricas']) if info_colunas['categoricas'] else 'Nenhuma'}
        - Colunas de data: {', '.join(info_colunas['datas']) if info_colunas['datas'] else 'Nenhuma'}

        Estatísticas básicas:
        {df.describe().to_string() if not df.empty else "Não disponível"}

        Insights identificados:
        {chr(10).join([f"- {insight}" for insight in insights]) if insights else "Nenhum insight significativo foi identificado."}

        Por favor, forneça um relatório detalhado com:
        1. Um resumo executivo dos dados
        2. Análise das principais variáveis e suas relações
        3. Interpretação dos insights identificados
        4. Recomendações práticas baseadas na análise
        5. Próximos passos sugeridos para análise mais aprofundada

        O relatório deve ser estruturado, direto e em linguagem adequada para tomada de decisão empresarial.
        """

        # Definir a URL e cabeçalhos para a API da OpenAI
        url = "https://api.openai.com/v1/chat/completions"
        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Bearer {api_key}"
        }

        # Definir o corpo da requisição
        data = {
            "model": "gpt-4o",
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.7,
            "max_tokens": 4000
        }

        print("Enviando dados para análise com a OpenAI...")

        # Enviar requisição para a API
        response = requests.post(url, headers=headers, json=data)

        # Verificar se a requisição foi bem-sucedida
        if response.status_code == 200:
            result = response.json()
            relatorio = result['choices'][0]['message']['content']
            print("\nRelatório gerado com sucesso!")
            return relatorio
        else:
            print(f"Erro na requisição: {response.status_code}")
            print(response.text)
            return f"Erro ao gerar relatório: {response.status_code} - {response.text}"

    except Exception as e:
        print(f"Erro ao gerar relatório com OpenAI: {e}")
        return f"Não foi possível gerar o relatório. Erro: {str(e)}"

def exportar_para_google_docs(relatorio, filename, autorizar_agora=True):
    """
    Exporta o relatório gerado para um novo documento do Google Docs

    Args:
        relatorio (str): O texto do relatório a ser exportado
        filename (str): Nome do arquivo original analisado (para título do documento)
        autorizar_agora (bool): Se True, abre o fluxo de autorização no momento

    Returns:
        str: URL do documento criado ou mensagem de erro
    """
    print("\n" + "="*80)
    print("EXPORTANDO RELATÓRIO PARA GOOGLE DOCS")
    print("="*80)

    try:
        # Instalar bibliotecas necessárias se não estiverem instaladas
        try:
            import gspread
            from google.auth import default
            from googleapiclient.discovery import build
            from google.oauth2.service_account import Credentials
            from google.colab import auth
        except ImportError:
            print("Instalando bibliotecas necessárias...")
            !pip install -q gspread google-auth google-api-python-client
            import gspread
            from google.auth import default
            from googleapiclient.discovery import build
            from google.oauth2.service_account import Credentials
            from google.colab import auth

        # Autenticar usando a conta Google conectada ao Colab
        if autorizar_agora:
            print("Autenticando no Google Drive...")
            auth.authenticate_user()

        # Obter credenciais autenticadas
        creds, _ = default()

        # Criar um serviço da API do Google Docs
        docs_service = build('docs', 'v1', credentials=creds)
        drive_service = build('drive', 'v3', credentials=creds)

        # Criar um novo documento
        titulo_doc = f"Análise de Dados - {filename} - {datetime.now().strftime('%d/%m/%Y %H:%M')}"
        doc_body = {
            'title': titulo_doc
        }

        print(f"Criando documento '{titulo_doc}'...")
        doc = docs_service.documents().create(body=doc_body).execute()
        document_id = doc.get('documentId')

        # Formatar o conteúdo para o documento
        conteudo = [
            {
                'insertText': {
                    'location': {
                        'index': 1
                    },
                    'text': relatorio
                }
            }
        ]

        # Inserir o conteúdo no documento
        docs_service.documents().batchUpdate(
            documentId=document_id,
            body={'requests': conteudo}
        ).execute()

        # Obter link para compartilhamento
        doc_url = f"https://docs.google.com/document/d/{document_id}/edit"

        # Atualizar permissões para que qualquer pessoa com o link possa visualizar
        permission = {
            'type': 'anyone',
            'role': 'reader',
            'allowFileDiscovery': False
        }

        drive_service.permissions().create(
            fileId=document_id,
            body=permission
        ).execute()

        print(f"Documento criado com sucesso!")
        print(f"Link para acesso: {doc_url}")

        return doc_url

    except Exception as e:
        print(f"Erro ao exportar para Google Docs: {e}")
        print("\nDica: Você pode copiar manualmente o relatório e colar em um novo documento do Google Docs.")
        return f"Não foi possível exportar para o Google Docs. Erro: {str(e)}"

def exportar_analise_para_pdf(df, filename, info_colunas, insights, relatorio_ia):
    """
    Exporta a análise completa para um arquivo PDF formatado

    Args:
        df (DataFrame): DataFrame analisado
        filename (str): Nome do arquivo original
        info_colunas (dict): Informações sobre as colunas
        insights (list): Lista de insights encontrados
        relatorio_ia (str): Relatório gerado pela IA

    Returns:
        str: Caminho do arquivo PDF gerado
    """
    print("\n" + "="*80)
    print("EXPORTANDO ANÁLISE COMPLETA PARA PDF")
    print("="*80)

    try:
        # Instalar a biblioteca FPDF se necessário
        try:
            from fpdf import FPDF
        except ImportError:
            print("Instalando biblioteca FPDF...")
            !pip install -q fpdf
            from fpdf import FPDF

        # Definir o nome do arquivo PDF de saída
        output_filename = f"Analise_{filename.split('.')[0]}_{datetime.now().strftime('%Y%m%d_%H%M')}.pdf"

        # Criar uma instância de PDF
        class PDF(FPDF):
            def header(self):
                # Logo
                # self.image('logo.png', 10, 8, 33)
                # Arial bold 15
                self.set_font('Arial', 'B', 15)
                # Mover para a direita
                self.cell(80)
                # Título
                self.cell(30, 10, 'Análise de Dados Automatizada', 0, 0, 'C')
                # Quebra de linha
                self.ln(20)

            def footer(self):
                # Posição a 1.5 cm do final
                self.set_y(-15)
                # Arial italic 8
                self.set_font('Arial', 'I', 8)
                # Número da página
                self.cell(0, 10, 'Página ' + str(self.page_no()) + '/{nb}', 0, 0, 'C')

            def chapter_title(self, title):
                # Arial 12
                self.set_font('Arial', 'B', 12)
                # Cor de fundo
                self.set_fill_color(200, 220, 255)
                # Título
                self.cell(0, 6, title, 0, 1, 'L', 1)
                # Quebra de linha
                self.ln(4)

            def chapter_body(self, body):
                # Times 12
                self.set_font('Times', '', 12)
                # Imprimir texto justificado
                self.multi_cell(0, 5, body)
                # Quebra de linha
                self.ln()

            def add_section(self, title, content):
                self.chapter_title(title)
                self.chapter_body(content)

        # Instanciar o objeto PDF
        pdf = PDF()
        pdf.alias_nb_pages()
        pdf.add_page()
        pdf.set_auto_page_break(auto=True, margin=15)

        # Informações gerais
        pdf.add_section('INFORMAÇÕES GERAIS',
                         f"Arquivo: {filename}\n"
                         f"Data da análise: {datetime.now().strftime('%d/%m/%Y %H:%M')}\n"
                         f"Número de linhas: {len(df)}\n"
                         f"Número de colunas: {df.shape[1]}")

        # Tipos de variáveis
        tipos_vars = (f"Variáveis numéricas: {', '.join(info_colunas['numericas']) if info_colunas['numericas'] else 'Nenhuma'}\n\n"
                      f"Variáveis categóricas: {', '.join(info_colunas['categoricas']) if info_colunas['categoricas'] else 'Nenhuma'}\n\n"
                      f"Variáveis de data: {', '.join(info_colunas['datas']) if info_colunas['datas'] else 'Nenhuma'}")
        pdf.add_section('TIPOS DE VARIÁVEIS', tipos_vars)

        # Estatísticas descritivas
        if len(info_colunas['numericas']) > 0:
            stats_desc = df[info_colunas['numericas']].describe().to_string()
            pdf.add_section('ESTATÍSTICAS DESCRITIVAS', stats_desc)

        # Insights
        if insights:
            insights_text = "\n\n".join([f"{i+1}. {insight}" for i, insight in enumerate(insights)])
            pdf.add_section('INSIGHTS IDENTIFICADOS', insights_text)

        # Relatório da IA
        if relatorio_ia:
            pdf.add_section('RELATÓRIO GERADO PELA IA', relatorio_ia)

        # Salvar o PDF
        pdf.output(output_filename)

        print(f"\nPDF gerado com sucesso: {output_filename}")

        # Se estiver no Google Colab, oferecer download do arquivo
        try:
            from google.colab import files
            files.download(output_filename)
            print(f"Download iniciado para {output_filename}")
        except:
            print(f"Arquivo salvo localmente em: {output_filename}")

        return output_filename

    except Exception as e:
        print(f"Erro ao exportar para PDF: {e}")
        return f"Não foi possível exportar para PDF. Erro: {str(e)}"
def analisar_planilha():
    """
    Função principal para executar toda a análise da planilha
    """
    print("="*80)
    print("ANÁLISE AUTOMÁTICA DE PLANILHAS COM IA")
    print("="*80)
    print("Este programa realiza uma análise exploratória automática de dados,")
    print("gerando visualizações, identificando insights e criando um relatório com OpenAI.")
    print("="*80)

    # Carregar os dados
    df, filename = carregar_dados()

    # Realizar análise exploratória
    print("\nRealizando análise exploratória...")
    info_colunas = analise_exploratoria(df)

    # Visualizar distribuições de variáveis numéricas
    if info_colunas['numericas']:
        visualizar_distribuicoes_numericas(df, info_colunas['numericas'])

    # Visualizar distribuições de variáveis categóricas
    if info_colunas['categoricas']:
        visualizar_distribuicoes_categoricas(df, info_colunas['categoricas'])

    # Visualizar séries temporais
    if info_colunas['datas']:
        visualizar_series_temporais(df, info_colunas['datas'], info_colunas['numericas'])

    # Visualizar correlações
    if len(info_colunas['numericas']) >= 2:
        visualizar_correlacoes(df, info_colunas['numericas'])

    # Realizar análise bivariada
    if info_colunas['numericas'] and info_colunas['categoricas']:
        analise_bivariada(df, info_colunas['numericas'], info_colunas['categoricas'])

    # Identificar clusters
    if len(info_colunas['numericas']) >= 2:
        identificar_clusters(df, info_colunas['numericas'])

    # Gerar insights
    insights = gerar_insights(df, info_colunas)

    # Gerar relatório com OpenAI
    relatorio = gerar_relatorio_openai(filename, df, info_colunas, insights)

    print("\n" + "="*80)
    print("RELATÓRIO DE ANÁLISE GERADO PELA IA")
    print("="*80)
    print(relatorio)

    # Exportar para PDF
    from IPython.display import display, HTML

    print("\n" + "="*80)
    print("OPÇÕES DE EXPORTAÇÃO")
    print("="*80)
    print("1. Exportar análise completa para PDF")
    print("2. Continuar sem exportar")

    resposta = input("\nSelecione uma opção (1-2): ")

    if resposta == '1':
        pdf_path = exportar_analise_para_pdf(df, filename, info_colunas, insights, relatorio)

    print("\n" + "="*80)
    print("ANÁLISE CONCLUÍDA")
    print("="*80)
    print(f"Nome do arquivo: {filename}")
    print(f"Número de linhas analisadas: {len(df)}")
    print(f"Número de insights encontrados: {len(insights)}")
    print("\nObrigado por usar o analisador automático de planilhas!")

    return df, info_colunas, insights, relatorio

# Executar a análise quando o script for executado
if __name__ == "__main__":
    analisar_planilha()

# Executar a análise quando o script for executado
if __name__ == "__main__":
    analisar_planilha()
```
